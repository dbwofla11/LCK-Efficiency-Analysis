솔로랭크 데이터에는 멀티킬, CC 시간 등 개인 교전 성향을 더 세밀하게 볼 수 있는 지표가 추가로 존재한다.

---

## 4. 딜량 / 피해 지표

| 의미 | 프로 경기 데이터 | 솔로랭크 데이터 |
|---|---|---|
| 분당 챔피언 딜량 | dpm | dpm |
| 팀 내 딜 비중 | damageshare | 없음 |
| 받은 피해량/분 | damagetakenperminute | damagetakenperminute |
| 감소시킨 피해량/분 | damagemitigatedperminute | damagemitigatedperminute |
| 총 챔피언 딜량 | 없음 | totalDamageDealtToChampions |
| 총 받은 피해량 | 없음 | totalDamageTaken |
| 총 감소 피해량 | 없음 | damageSelfMitigated |

### 해석

`dpm`, `damagetakenperminute`, `damagemitigatedperminute`는 공통으로 사용할 수 있다.  
다만 `damageshare`는 프로 데이터에만 있으므로 솔로랭크와 직접 비교하기 어렵다.

---

## 5. 시야 / 운영 지표

| 의미 | 프로 경기 데이터 | 솔로랭크 데이터 |
|---|---|---|
| 시야 점수 | visionscore | visionScore |
| 분당 시야 점수 | vspm | vspm |
| 와드 설치 | wardsplaced | wardsPlaced |
| 와드 제거 | wardskilled / wardskilled 오타 주의 | wardsKilled |
| 분당 와드 설치 | wpm | wpm |
| 분당 와드 제거 | wcpm | wcpm |
| 제어 와드 구매 | controlwardsbought | visionWardsBoughtInGame |

### 해석

시야 관련 지표는 두 데이터 모두 잘 대응된다.  
프로 기준과 솔로랭크 기준을 비교하기 좋은 영역이다.

## 6. 경제 / 성장 지표

| 의미 | 프로 경기 데이터 | 솔로랭크 데이터 |
|---|---|---|
| 획득 골드 | earnedgold | goldEarned |
| 분당 골드 | earned gpm | earned_gpm |
| 팀 골드 비중 | earnedgoldshare | 없음 |
| 분당 CS | cspm | cspm |
| 총 CS | 없음 | cs |
| 챔피언 레벨 | 없음 | champLevel |

### 해석

`earnedgold`, `earned gpm`, `cspm`은 핵심 공통 지표로 사용할 수 있다.  
다만 프로 데이터의 `earnedgoldshare`는 팀 내 자원 점유율이라 솔로랭크와 직접 대응되는 값은 없다.

---

## 7. 오브젝트 / 타워 지표

| 의미 | 프로 경기 데이터 | 솔로랭크 데이터 |

| 바론 킬 | 없음 | baronKills |

### 해석

타워 피해량은 두 데이터에서 비교 가능하다.  
솔로랭크 데이터에는 오브젝트 관련 지표가 더 많이 포함되어 있어 운영 성향 분석에 추가 활용할 수 있다.

# 최종 정리

현재 데이터는 다음과 같이 사용할 수 있다.

## 1. 공통 Feature 기반 비교

프로 데이터와 솔로랭크 데이터에서 공통 의미를 갖는 feature를 사용한다.

예:

- dpm
- cspm
- vspm
- wpm
- wcpm
- earned_gpm
- damagetakenperminute
- damagemitigatedperminute

---

## 2. 프로 전용 Feature는 기준선 분석에 사용

라인전 15분 지표나 팀 내 비중 지표는 프로 데이터에만 존재하므로  
프로 선수 유형 분석 또는 구단 스타일 분석에 사용한다.

---

## 3. 솔로랭크 전용 Feature는 지망생 세부 진단에 사용

멀티킬, 챔피언, 오브젝트 피해량 등은  
솔로랭크 내부에서 지망생의 성향을 더 자세히 진단하는 데 사용할 수 있다.
